```python
import numpy as np

import matplotlib.pyplot as plt

from keras.models import Sequential
from keras.layers import Dense
from keras import backend as K
from keras import optimizers

%matplotlib inline

# Configure the plot style
params = {'axes.labelsize': 20, 'xtick.labelsize': 20, 'ytick.labelsize': 20,
          'figure.figsize': (8,8), 'text.usetex': True,
          'font.family': 'FreeSerif'}
plt.rcParams.update(params)
```

```python
# Define some functions

def A(x):
    '''
        Left part of initial equation
    '''
    return x + (1. + 3.*x**2) / (1. + x + x**3)


def B(x):
    '''
        Right part of initial equation
    '''
    return x**3 + 2.*x + x**2 * ((1. + 3.*x**2) / (1. + x + x**3))


def f(x, psy):
    '''
        d(psy)/dx = f(x, psy)
        This is the f() function on the right
    '''
    return B(x) - psy * A(x)


def psy_analytic(x):
    '''
        Analytic solution of the DE
    '''
    return (np.exp((-x**2)/2.)) / (1. + x + x**3) + x**2
```

```python
# Define the loss function

def loss_function(network_output,network_input):

    def loss(y_true, y_pred):
        
        x = network_input
        N = network_output
        
        psy_t = 1. + x*N
        
        d_N = K.gradients(N,x)
        
        d_psy_t = N + x*d_N

        err_sqr = (d_psy_t - f(x,psy_t))**2

        return K.sum(err_sqr)

    return loss
```

```python
x_vals = np.linspace(0, 10, 5000) 
y_vals = psy_analytic(x_vals)
```

```python
model = Sequential()
model.add(Dense(50,input_shape=(1,),activation='relu'))
model.add(Dense(100,activation='relu'))
model.add(Dense(100,activation='relu'))
model.add(Dense(50,activation='relu'))
model.add(Dense(1,activation='linear'))
```

```python
# The y_true values aren't used in the loss function, so we just pass a dummy array as y_true
model.compile(loss=loss_function(model.output,model.input), optimizer=optimizers.Adam(lr=1e-3))
model.fit(x_vals,np.zeros(x_vals.shape),epochs=200,batch_size=64)
model.compile(loss=loss_function(model.output,model.input), optimizer=optimizers.Adam(lr=5e-4))
model.fit(x_vals,np.zeros(x_vals.shape),epochs=200,batch_size=64)
model.compile(loss=loss_function(model.output,model.input), optimizer=optimizers.Adam(lr=1e-4))
model.fit(x_vals,np.zeros(x_vals.shape),epochs=200,batch_size=64)
model.compile(loss=loss_function(model.output,model.input), optimizer=optimizers.Adam(lr=5e-5))
model.fit(x_vals,np.zeros(x_vals.shape),epochs=200,batch_size=64)
```

```python
y_pred = 1. + x_vals*model.predict(x_vals).reshape(x_vals.shape)
```

```python
fig = plt.figure(1, facecolor='white')
ax = fig.add_axes([0.15, 0.15, 0.9, 0.6])
ax.plot(x_vals,y_vals,label=r'$\mathrm{Analytic}$')
ax.plot(x_vals,y_pred,label=r'$\mathrm{Predicted}$')
ax.set_xlabel(r'$x$')
ax.set_ylabel(r'$\Psi(x)$')
ax.minorticks_on()
ax.legend(frameon=False,loc='best')
plt.show()
```
